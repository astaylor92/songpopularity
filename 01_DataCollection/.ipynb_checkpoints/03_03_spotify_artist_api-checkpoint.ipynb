```python
uri_artist = pd.read_csv('uri_artist.csv')[['sp_artist_uri', 'sp_uri']]
uri_artist = uri_artist.values.tolist()
uri_artist[:4]
```

```python
CLIENT_ID = '75855396ac0446279df54c8a7df71414'
CLIENT_SECRET = '7bc71c24b81743ebbb79f0b612c26587'
```

```python
AUTH_URL = 'https://accounts.spotify.com/api/token'

# POST
auth_response = requests.post(AUTH_URL, {
    'grant_type': 'client_credentials',
    'client_id': CLIENT_ID,
    'client_secret': CLIENT_SECRET,
})

# convert the response to JSON
auth_response_data = auth_response.json()

# save the access token
access_token = auth_response_data['access_token']
```

```python
headers = {
    'Authorization': 'Bearer {token}'.format(token=access_token)
}
```

```python
# Initate to start
artisturi_artistinfo = []
j = 0
```

```python
# Initiate from file
# isrc_uri_audiofeats = pd.read_csv('isrc_uri_audiofeats.csv')[[TBD]]
# isrc_uri_audiofeats = isrc_uri.values.tolist()
# isrc_uri_audiofeats = TBD
# j = len(isrc_uri_audiofeats)
```

```python
# actual GET request with proper header
for j in range(j, int(np.ceil(len(uri_artist)/50))):
    artist_uris = ','.join([artist_uri for track_uri, artist_uri in uri_artist[j*50:(j+1)*50]])
    
    retries = 1
    success = False
    while not success:
        try:
            r = requests.get(BASE_URL + 'artists?ids='+artist_uris, headers=headers)
            success = True
        except Exception as e:
            wait = retries * 30;
            print(f'Error! Waiting {wait} secs and re-trying...')
            sys.stdout.flush()
            time.sleep(wait)
            retries += 1
    
    resp_header = r.headers
    data = r.json()['artists']
    i = 0
    for artist in data:
        if artist is not None:
            artisturi_artistinfo.append([artist['uri'], artist['popularity'], artist['genres'], artist['followers']['total']])
        i += 1
    time.sleep(np.random.random()*2)
    j += 1
    print("{} uris searched, {} artists found".format(j*50, len(artisturi_artistinfo)))
    if j % 1000 == 0:
        # POST
        auth_response = requests.post(AUTH_URL, {
            'grant_type': 'client_credentials',
            'client_id': CLIENT_ID,
            'client_secret': CLIENT_SECRET,
        })

        # convert the response to JSON
        auth_response_data = auth_response.json()

        # save the access token
        access_token = auth_response_data['access_token']
        
        headers = {
            'Authorization': 'Bearer {token}'.format(token=access_token)
        }
```
